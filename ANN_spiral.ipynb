Import, split, and normalize Data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
data = pd.read_csv('3dSinusoidalANN.csv')
X = data[['x','z']]
y = data['y']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state= 42)
#X_train, X_valid, y_train, y_valid = train_test_split(X_train,y_train, random_state=42)

train_df = X_train.merge(y_train, left_index= True, right_index= True).reset_index(drop = True)
test_df = X_test.merge(y_test, left_index=True, right_index=True).reset_index(drop=True)
scaler = StandardScaler()

train_df = pd.DataFrame(scaler.fit_transform(train_df)).rename(columns = {0:'x',
                                                                            1:'z',
                                                                         2:'y'
}).sort_values(by ='z')


test_df = pd.DataFrame(scaler.transform(test_df)).rename(columns = {0:'x',
                                                                      1:'z',
                                                                        2:'y'
}).sort_values(by ='z')


X_train = train_df[['x','z']]
y_train = train_df['y']
```

# Explore 3D Sinusoidal Data with Artifical Neural Networks

Fitting a `Sequential` model to this data.

```python
X_train.shape, y_train.shape, train_df.shape
```

```python
keras.backend.clear_session()
np.random.seed(42)
tf.random.set_seed(42)
model = keras.models.Sequential([
    keras.layers.Dense(1000, input_shape=X_train.shape[1:],activation ='relu'),
    keras.layers.Dense(1000, input_shape=X_train.shape[1:], activation = 'relu'),
    keras.layers.Dense(1000, input_shape=X_train.shape[1:],activation = 'relu'),
    keras.layers.Dense(1000, input_shape=X_train.shape[1:], activation = 'relu'),
     keras.layers.Dense(500, input_shape=X_train.shape[1:], activation = 'relu'),
    keras.layers.Dense(1, activation = 'linear')
])
opt = keras.optimizers.SGD(learning_rate=0.05)
model.compile(loss='mean_squared_error', optimizer=opt)

history = model.fit(X_train, y_train, epochs =100,
                    #validation_data = (X_valid, y_valid),
                    ) 
```

Use the model's `predict()` method to make a prediction for `y` using the `x` and `z` training data.  Use the `plotscatter3Ddata(fit_x, fit_y, fit_z, scat_x, scat_y, scat_z)` function to plot the data and the prediction curve.
